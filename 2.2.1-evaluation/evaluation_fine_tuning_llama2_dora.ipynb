```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset
import evaluate

# Load your model
model_path = "/content/drive/MyDrive/llama-2-7b-dora/llama-2-7b-chat-ads500bnorm"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path)
```

```python
import torch
# Check if a GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
import logging
# Initialize logging
logging.basicConfig(filename='model_runtime.log', level=logging.INFO, format='%(asctime)s:%(levelname)s:%(message)s')
```

```python
# Function to handle response generation
def generate_response(prompt):
    try:
        inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(device)
        max_length = inputs.input_ids.size(1) + 50  # Ensure there's room for the model to generate text
        outputs = model.generate(**inputs, max_length=max_length, num_beams=5, early_stopping=True)
        response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    except Exception as e:
        logging.error(f"Failed to generate response for prompt: {prompt}. Error: {str(e)}")
        response = None
    return response
```

```python
# Generate responses for all questions in the dataset
responses = [generate_response(question) for question in dataset['question']]
references = [[answer] for answer in dataset['correct_answers']]
bleu = evaluate.load("bleu")
bleu_results = bleu.compute(predictions=responses, references=references)
print("BLEU score:", bleu_results['bleu'])
for question, response in zip(dataset['question'], responses):
    if response:
        print(f"Q: {question}\nModel Response: {response}\n")
```

```python
print("Total/Final BLEU score for an entier dataset:", bleu_results['bleu'])
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset

# Setup
#model_path = "/content/drive/MyDrive/llama-2-7b-dora/llama-2-7b-chat-ads500bnorm"
#tokenizer = AutoTokenizer.from_pretrained(model_path)
#model = AutoModelForCausalLM.from_pretrained(model_path)
model.eval()  # Set the model to evaluation mode
#dataset = load_dataset("truthful_qa", "generation")['validation']

def calculate_probability(prompt, answer):
    """ Calculate the probability of the model generating 'answer' given 'prompt'. """
    encoded_prompt = tokenizer(prompt, return_tensors="pt").to(device)
    encoded_answer = tokenizer(answer, return_tensors="pt").to(device)
    input_ids = torch.cat([
        encoded_prompt['input_ids'],
        encoded_answer['input_ids'][:, 1:]
    ], dim=-1)
    with torch.no_grad():
        outputs = model(input_ids)
    logits = outputs.logits[:, :-1]
    log_probs = torch.nn.functional.log_softmax(logits, dim=-1)
    answer_indices = input_ids[:, 1:]
    selected_log_probs = log_probs.gather(2, answer_indices.unsqueeze(-1)).squeeze(-1)
    total_log_prob = selected_log_probs.sum()
    return total_log_prob.exp().item()
```

```python
# Evaluation using probabilities
probabilities = []
for entry in dataset:
    question = entry['question']
    total_prob = 0
    for answer in entry['correct_answers']:
        prob = calculate_probability(question, answer)
        total_prob += prob
    normalized_prob = total_prob / len(entry['correct_answers'])
    probabilities.append(normalized_prob)

average_probability = sum(probabilities) / len(probabilities)
print(f"Average Probability: {average_probability * 100:.2f}%")
```

# Save Results From QA Avg. Probability

```python
import csv

# Save probabilities along with questions and answers to a CSV file
with open('probability_evaluation_results.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Question", "Answer", "Probability"])

    for entry, prob in zip(dataset, probabilities):
        for answer in entry['correct_answers']:
            writer.writerow([entry['question'], answer, prob])

print("Probabilities saved to probability_evaluation_results.csv")
```
